```python
import numpy as np
a = np.array([-1, 0, 1])
b = np.array([2, -1, 1])
np.dot(a,b)
```

```python
a + b #ベクトル和
```

```python
al = [-1, 0, 1]
bl = [2, -1, 1]
al + bl
```

```python
#cos類似度
np.dot(a,b)/(np.linalg.norm(a) * np.linalg.norm(b))
```

```python
#cos類似度を返す関数
def getCos(a, b):
  return np.dot(a,b)/(np.linalg.norm(a) * np.linalg.norm(b))
```

```python
getCos(a, b)
```

```python
a1 = np.array([1, 1, 2, 0])
b1 = np.array([-1, 2, 0.5, 1])
getCos(a1,b1)
```

```python
A = np.array([[2,3],[3,-2]])
b = np.array([-4, 7])
A_inv = np.linalg.inv(A)
A_inv
```

```python
x = np.dot(A_inv, b)
x
```

```python
np.linalg.solve(A, b)
```

```python
A = np.array([[5,3],[4,9]])
eig = np.linalg.eig(A)
print(eig[0])
print(eig[1])
```

```python
P = eig[1]
D = np.linalg.inv(P) @ A @ P
D
```
